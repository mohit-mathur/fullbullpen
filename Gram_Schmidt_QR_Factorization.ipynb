# Problem 4
### Part (a)

```python
import numpy as np
import time

#define a function to compute QR decomposition of a matrix A. 

def qr_factorization(A):
    m,n = A.shape  

    #initialise the matrices Q and R 
    Q = np.zeros((m,n)) 
    R = np.zeros((n,n))
    for j in range(n):
      a = A[: , j]
    
      # apply the Gram-Schmidt Process
      for i in range(j):
        q = Q[: , i] 
        R[i , j] = np.dot(q,a)
        a = a - R[i , j] * q 
      
      #normalise the vectors    
      norm = np.linalg.norm(a)  
      Q[: ,j] = a / norm
      R[j , j] = norm
    return Q, R  
  
```

## Part (c)

```python

""" Check that above code yields correct result for the matrix above,
and print Q, R and A−QR. """

A = np.array([[12, -51, 4],
     [6, 167, -68],
     [-4, 24, -41]])
Q, R = qr_factorization(A)

print('Q = ',Q,"\n")
print('R = ',R,"\n")
W = np.dot(Q,R)
print('Q * R = ',W,"\n")
print('A − Q*R = ', A-W)
```

##  Part (d)

```python

#Compute QR factorization for matrices with N = 10, 100, 500, 1000 and 2000 
#and calculate the time taken for each.

# For N = 10
A10 = np.random.rand(10, 10)

start10 = time.time()
Q10, R10 = qr_factorization(A10)
end10 = time.time()

time_N10 = (end10 - start10)

print('Q10 = ',Q10)
print('R10 = ',R10)

W10 = np.dot(Q10,R10)
print('A10 = ',W10)

# For N = 100
A100 = np.random.rand(100, 100)
start100 = time.time()
Q100, R100 = qr_factorization(A100)
end100 = time.time()
time_N100 = (end100 - start100)

print('Q100 = ',Q100)
print('R100 = ',R100)

W100 = np.dot(Q100,R100)
print('A100 = ',W100)

#For N = 500
A500 = np.random.rand(500, 500)
start500 = time.time()
Q500, R500 = qr_factorization(A500)
end500 = time.time()
time_N500 = (end500 - start500)

print('Q500 = ',Q500)
print('R500 = ',R500)

W500 = np.dot(Q500,R500)
print('A500 = ',W500)

# For N = 1000
A1000 = np.random.rand(1000, 1000)
start1000 = time.time()
Q1000, R1000 = qr_factorization(A1000)
end1000 = time.time()
time_N1000 = (end1000 - start1000)

print('Q1000 = ',Q1000)
print('R1000 = ',R1000)

W1000 = np.dot(Q1000,R1000)
print('A1000 = ',W1000)

# For N = 2000
A2000 = np.random.rand(2000, 2000)
start2000 = time.time()
Q2000, R2000 = qr_factorization(A2000)
end2000 = time.time()
time_N2000 = (end2000 - start2000)

print('Q2000 = ',Q2000)
print('R2000 = ',R2000)

W2000 = np.dot(Q2000,R2000)
print('A2000 = ',W2000)
```

```python

print("Time for 10x10 matrix: ",time_N10)
print("Time for 100x100 matrix: ",time_N100)
print("Time for 500x500 matrix: ",time_N500)
print("Time for 1000x1000 matrix: ",time_N1000)
print("Time for 2000x2000 matrix: ",time_N2000)

time_qr = {10:time_N10,100:time_N100,500:time_N500,1000:time_N1000,2000:time_N2000 }

```

```python
import matplotlib.pylab as plt

# plot the graph of N vs Time 

list1 = sorted(time_qr.items())
x,y = zip(*list1)
plt.plot(x, y, 'm')
plt.xlabel("N")
plt.ylabel("Time (sec)")
plt.title('QR Factorization')

plt.show()
```

```python
# plot the graph of N vs Time in log-log scale

list2 = sorted(time_qr.items())
x,y = zip(*list1)
plt.loglog(x, y, 'r')
plt.xlabel("N")
plt.ylabel("Time (sec)")
plt.title("QR Factorization log-log Scale")
plt.show()
```
